```python
統計碼表 = 迩原碼表.統計(False)
末序 = -2
寡文 = [統計碼表["文"].most_common()[-1]]
while 統計碼表["文"].most_common()[末序][1] == 寡文[0][1]:
    寡文.append(統計碼表["文"].most_common()[末序])
    末序 -= 1
print(f'凡 {len(統計碼表["文"])} 个文。\n最頻：{統計碼表["文"].most_common(10)}\n最寡 {len(寡文)} 个，每文 {寡文[0][1]} 現：')
寡文 = list(zip(*寡文))[0]
print('、'.join(寡文))
if len(統計碼表["非文"]) > 0:
    print(f'非正解中出現 {len(統計碼表["非文"])} 个字榦，請繼續解之：')
    print(統計碼表["非文"])
```
